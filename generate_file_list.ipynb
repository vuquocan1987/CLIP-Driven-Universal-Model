```python
import os
import sys

data_set_path = "/home/an/Workspace/data/MOTS_CLIP/"
```

```python
# generate the txt file in this form:
# 04_LiTS/img/liver_16.nii.gz	04_LiTS/label/liver_16.nii.gz
# where the first is the image and the second is the label


IMAGE_FOLDER = "imagesTr"
LABEL_FOLDER = "labelsTr"

```

```python
sub_dataset = os.listdir(data_set_path)
output_path = "dataset/dataset_list/UNISEG.txt"
__file__ = os.getcwd()
output_path = os.path.join(os.path.dirname(__file__), output_path)
output_path = "/home/an/Workspace/CLIP-Driven-Universal-Model/dataset/dataset_list/UNISEG.txt"
for sub in sub_dataset:
    image_folder = os.path.join(data_set_path, sub, IMAGE_FOLDER)
    label_folder = os.path.join(data_set_path, sub, LABEL_FOLDER)
    image_files = os.listdir(image_folder)
    label_files = os.listdir(label_folder)
    label_files.sort()
    image_files.sort()
    if len(label_files) != len(image_files):
        print("Number of files not equal")
        sys.exit(1)
    with open(output_path, "a") as f:
        for i in range(len(label_files)):
            im_path = os.path.join(sub, IMAGE_FOLDER, image_files[i])
            lb_path = os.path.join(sub, LABEL_FOLDER, label_files[i])
            f.write(f"{im_path}\t{lb_path}\n")    

# sorts the files and check if number of files are equal, and write to the output file
        
```
