```python
import os
os.getcwd()
os.chdir('/app')
from config import MAPPING_MOTS_UNI_CLIP,SIMPLE_MAPPING
```

```python
# make directory for MOTS data for clip

data_path = '/app/data'
clip_path = os.path.join(data_path, 'Clip')
# make directory for MOTS data for clip
os.makedirs(clip_path, exist_ok=True)
path_mots_uniseg = os.path.join(data_path, 'MOTS_formatted')
for key in MAPPING_MOTS_UNI_CLIP:
    os.symlink(os.path.join(path_mots_uniseg, key), os.path.join(clip_path, MAPPING_MOTS_UNI_CLIP[key]))
```

```python
def generate_data_file_mapping(mapping,data_mapping_path="dataset/dataset_list/MOTS.txt"):
    with open(data_mapping_path, 'w') as f:
        for key in mapping:
            train_files = [f for f in os.listdir(os.path.join(clip_path, mapping[key],'imagesTr')) if os.path.isfile(os.path.join(clip_path, mapping[key],'imagesTr', f))]
            train_files.sort()
            for train_file in train_files:
                if mapping[key] == "04LiTS":
                    label_file = "segmentation"+train_file[5:]
                else:
                    label_file = train_file
                if os.path.exists(os.path.join(clip_path, mapping[key],'labelsTr', label_file)):
                    f.write(f"{os.path.join(mapping[key], 'imagesTr',train_file)} {os.path.join(mapping[key], 'labelsTr', label_file)}\n")

generate_data_file_mapping(MAPPING_MOTS_UNI_CLIP)
generate_data_file_mapping(SIMPLE_MAPPING, "dataset/dataset_list/MOTS_simple.txt")
```
